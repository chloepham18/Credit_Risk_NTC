# üìä NTC Consumer Credit: Credit Risk Modeling for New-To-Credit (NTC) Customers

This project focuses on building a credit risk model for New-To-Credit (NTC) customers to predict the likelihood of charge-offs, enabling financial institutions to make informed, risk-based decisions when approving credit applications. The model prioritizes **precision** (0.667 on validation) to support growth while minimizing default rates.

---

## üîç Project Overview

- **Goal**: Assess credit risk for NTC customers and develop a risk-based underwriting strategy.
- **Target**: Predict `Ever_ChargeOff` (binary) using a Random Forest Classifier.
- **Focus**: High precision (threshold 0.7) to minimize false approvals of risky customers.
- **Key Metrics**:
  - Precision: 0.667 (Validation), 0.53 (Test)
  - AUC: 0.953 (Validation), 0.94 (Test)

## üí° Recommendations for NTC Customer Approvals

Based on the analysis and the model‚Äôs precision-focused strategy (threshold 0.7), we recommend approving NTC customers with the following profiles:

- **FICO Score 600+ (preferably 640+)**: These customers (12.38% in 640+) are lower risk, especially with assets in the $501-1500 range.
- **Asset Band $501-1500**: This range (14.49%-16.83%) indicates financial stability.
- **DTI <3%**: Customers with DTI <1% (15.31%) or 1-3% (14.24%) in FICO 640+ or asset band $501-1500 are low-risk.
- **Loan Amount <10K**: Approve customers with loans <10K (14.20%), especially in FICO 640+ (13.11%) or asset band $501-1500 (18.88%).
- **Utilization <30%**: Very low (<10%) or low (10-30%) utilization in FICO 640+ (15.45%) or asset band $501-1500 (16.47%) indicates responsible credit use.

### Profiles to Avoid:
- **FICO <520**: High risk (19.44%), especially with assets <500 (29.41%) or high utilization (50.00%).
- **High DTI (7-10%)**: Particularly in FICO <520 or asset band <500 (22.47%).
- **High Loan Amounts (>10K-25K)**: Risky in FICO <520 (24.14%) or asset band <500 (15.73%).
- **High Utilization (60-100%)**: Avoid in FICO <520 (50.00%) or asset band <500 (17.57%).
